Построим гистограммы распределения заказов и обращений по неделям

```python
plt.figure(figsize=(15, 10))
df["Заказы"].plot(kind='bar')
plt.title('Распределение заказов по неделям')
plt.xlabel('Неделя')
plt.ylabel('Количество заказов')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
df["Обращения"].plot(kind='bar')
plt.title('Распределение обращений по неделям')
plt.xlabel('Неделя')
plt.ylabel('Количество обращений')
plt.show()
```

Также посмотрим на корреляцию между заказами и обращениями за первые 20 недель

```python
np.corrcoef(df["Обращения"][0:16], df["Заказы"][0:16])
```

Данные сильно коррелируеют, практически линейная зависиомость

Воспользуемся моделью регрессии Ridge

```python
model = Ridge()
```

```python
X = np.array(df["Заказы"]).reshape(-1, 1)
```

```python
y = np.array(df["Обращения"])
```

```python
model.fit(X[0:16], y[0:16])
```

```python
pred = model.predict(X[16:52])
```

Обновим наши данные с учетом обучения модели

```python
res = df.copy()
for i in range(36):
    res["Обращения"][i+17] = pred[i]
```

```python
plt.figure(figsize=(15, 10))
res["Обращения"].plot(kind='bar')
plt.title('Распределение обновленных обращений по неделям')
plt.xlabel('Неделя')
plt.ylabel('Обращения')
plt.show()
```

# Необходимое количество людей для покрытия нагрузки

Поскольку за час оператор может обработать 25 чатов или 20 звонков, то за рабочую неделю 40 часов полцчатся следующие значения:

```python
chats = 25*40
calls = 20*40
```

```python
chats
```

```python
calls
```

Переведем все обращения в обращения по чатам(45%) и по звонкам(35%) и посчитаем число операторов, необходимое для их обработки

```python
support_chats = [el/chats for el in res["Обращения"]*0.45]
support_calls = [el/calls for el in res["Обращения"]*0.35]
support = [math.ceil(i + j) for i,j in zip(support_calls, support_chats)]
```

```python
res.insert(2, "Операторы", support)
```

```python
plt.figure(figsize=(15, 10))
res["Операторы"].plot(kind='bar')
plt.title('Распределение необходимого числа операторов по неделям')
plt.xlabel('Неделя')
plt.ylabel('Операторы')
plt.show()
```

- В недели максимальной нагрузки необходимо 4 оператора
- В недли с низкой нагрузкой достаточно 2 операторов
- 31 неделя нагружена, поэтому нужно 3 оператора
- Чтобы справиться с нагрузкой в 44 неделю необходимо 2 оператора
